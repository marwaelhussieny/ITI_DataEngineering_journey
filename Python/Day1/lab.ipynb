```python
num = int(input("enter a number: "))

if num <= 1:
    print("Not prime ")
else:
    prime = True
    for i in range(2, num):
        if num % i == 0:
            prime = False
            break

    if prime:
        print("Prime number")
    else:
        print("Not a prime number")


        
"""num = int(input("Enter a number: "))

if num <= 1:
    print("Not a prime number")

elif num == 2:
    print("Prime number")

elif num % 2 == 0:
    print("Not a prime number")

elif num == 3:
    print("Prime number")

elif num % 3 == 0:
    print("Not a prime number")

elif num == 5:
    print("Prime number")

elif num % 5 == 0:
    print("Not a prime number")

else:
    print("Prime number")
        """
```

11- Write a program that asks the user for:

 - The total number of students enrolled in a course

 - The desired number of students per group

The program should then calculate and display the total number of groups formed. If the students cannot be divided evenly, one group may contain fewer members than specified.

**Example output (1):**

- How many students on the course? 8
- Desired group size? 4
- Number of groups formed: 2


**Example output (2):**

- How many students on the course? 11
- Desired group size? 3
- Number of groups formed: 4


> Hint: The integer division operator // may be helpful in solving this problem.

```python
students = int(input("enter number of students "))
group_size = int(input("enter desired group size "))

groups = (students + group_size - 1) // group_size
print(f"Number of groups formed: {groups}")

                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                      
```
