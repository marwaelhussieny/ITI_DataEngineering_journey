### **1. squares numers:**
- Create a dictionary where keys are the first 5 natural numbers and values are their squares.
- Write a script to:
  - Print all keys and their respective values.
  - Add a new key-value pair: `6: 36`.
  - Check if a specific key (e.g., 4) exists in the dictionary.


```python
squares = {
    1: 1,
    2: 4,
    3: 9,
    4: 16,
    5: 25}
for key, value in squares.items():
    print(key, ":", value)

squares[6] = 36

if 4 in squares:
    print("Key 4 exists")
else:
    print("Key 4 does not exist")
```

### **2. Student Grades:**
- Create a dictionary to store student names as keys and a list of their grades as values.
- Write a script to:
  - Calculate the average grade for each student.
  - Find the student with the highest average.


```python
students = {
    "Ali": [80, 90, 85],
    "Sara": [95, 88, 92],
    "Omar": [70, 75, 78]
}

averages = {}

for name, grades in students.items():
    avg = sum(grades) / len(grades)
    averages[name] = avg
    print(name, "average:", avg)

best_student = max(averages, key=averages.get)
print("Highest average:", best_student)
```


### **4. Employee Database:**
- Create a list of dictionaries where each dictionary represents an employee with attributes:
  - `name`, `age`, and `department`.
- Write a script to:
  - Filter employees by a specific department.
  - Find the average age of employees.


```python
employees = [{"name": "Ali", "age": 30, "department": "IT"},{"name": "Sara", "age": 25, "department": "HR"},{"name": "Omar", "age": 35, "department": "IT"}]

dept = input("Enter department to filter: ")

found = False
for emp in employees:
    if emp["department"].lower() == dept.lower(): 
        print(emp)
        found = True

if not found:
    print("No employees found in this department")


total_age = 0
for emp in employees:
    total_age += emp["age"]

average_age = total_age / len(employees)
print("Average age:", average_age)
```
