### **1) Calculator**

Please write a program which asks the user for two numbers and an operation. If the operation is add, multiply or subtract, the program should calculate and print out the result of the operation with the given numbers. If the user types in anything else, the program should print out nothing.

*Some examples of expected behaviour:*

Number 1: 10<br>
Number 2: 17<br>
Operation: add<br>

10 + 17 = 27

Number 1: 4<br>
Number 2: 6<br>
Operation: multiply<br>

4 * 6 = 24

```python
num1 = int(input("enter Number 1"))
num2 = int(input("enter number 2"))
operation = input("enter operation ")

if operation == "add":
    print(f"{num1} + {num2} = {num1 + num2}")
elif operation == "subtract":
    print(f"{num1} - {num2} = {num1 - num2}")
elif operation == "multiply":
    print(f"{num1} * {num2} = {num1 * num2}")
```

### **2) Daily wages**

Please write a program which asks for the hourly wage, hours worked, and the day of the week. The program should then print out the daily wages, which equal hourly wage multiplied by hours worked, except on Sundays when the hourly wage is doubled.

Hourly wage: 8.5<br>
Hours worked: 3<br>
Day of the week: Monday<br>
Daily wages: 25.5 euros<br>

Hourly wage: 12.5<br>
Hours worked: 10<br>
Day of the week: Sunday<br>
Daily wages: 250.0 euros<br>

```python
wage = float(input(" enter hourly wage "))
hours = float(input("enter hours worked: "))
day = input("enter day of the week: ")

if day == "Sunday":
    wage *= 2

daily_wage = wage * hours
print(f"Daily wages: {daily_wage} euros")
```

### **3) PIN and number of attempts**

Please write a program which keeps asking the user for a PIN code until they type in the correct one, which is 4321. The program should then print out the number of times the user tried different codes.

PIN: 3245<br>
Wrong<br>
PIN: 1234<br>
Wrong<br>
PIN: 0000<br>
Wrong<br>
PIN: 4321<br>
Correct! It took you 4 attempts

**If the user gets it right on the first try,** the program should print out something a bit different:

PIN: 4321<br>
Correct! It only took you one single attempt!

```python
correct_pin = "4321"
attempts = 0

while True:
    pin = input("enter pin: ")
    attempts += 1
    
    if pin == correct_pin:
        if attempts == 1:
            print ("right with single attempt")
            break
        else:
            print (f"right pin with {attempts} tries")
            break
    else:
        print("try again")


```

### **5) write a function which has an input of string from user then it return the same string reversed.**

```python
##slicing
def reverse_string(st):
    return st[::-1]

user_input = input("Enter a string: ")
print(reverse_string(user_input))


#-----------------loop
def reverse_stringg(string):
    reversed_str = ""
    for char in string:
        reversed_str = char + reversed_str
    return reversed_str

user_inputt = input("Enter a string: ")
print(reverse_stringg(user_inputt))
#---------------------reserved function
def reverse_stringgg(s):
    return ''.join(reversed(s))

user_inputtt = input("Enter a string: ")
print(reverse_stringgg(user_inputtt))

```

7) Write a function sum_series(n) that calculates the sum of the 
series: 
1 - 2 + 3 - 4 + 5 - ... up to n

```python
def sum_series(n):
    result = 0

    for number in range(1, n + 1):
        if number % 2 == 0:
            result = result - number
        else:
            result = result + number

    return result
```

8) Write a Python function to calculate the factorial of a number (a 
non-negative integer). The function accepts the number as an 
argument.

```python
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
```
