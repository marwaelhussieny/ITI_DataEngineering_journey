
## File I/O

**Task 1:**
Write a program that:
- Takes user input for **name**, **age**, **email** and **track**.
- Raises the exception if the age is **less than 20**.
    - valueError if the age is not an integer.
- Validates the inputs to ensure correctness as follows:
    - Name and track must only contain alphabetic characters and spaces.
    - Age must be a positive integer.
    - Email must follow a standard email format (e.g., contain "@" and ".").
- Writes the data to a file called **`users.txt`** in the format:
        `"Name: John, Age: 25, Email:user@gmail.com Track: Power BI"`
- Read the content from **`users.txt`**  and print each line in **uppercase**.
- Counts the number of words and Prints the **total count**.
- catch any exception occurred and display a **custom error message**.
---

```python
try:
    name = input("Enter name: ")
    age = input("Enter age: ")
    email = input("Enter email: ")
    track = input("Enter track: ")

    if not age.isdigit():
        raise ValueError("Age must be a number")

    age = int(age)

    if age < 20:
        raise Exception("Age must be at least 20")

    if not name.replace(" ", "").isalpha():
        raise Exception("Name must contain letters only")

    if not track.replace(" ", "").isalpha():
        raise Exception("Track must contain letters only")

    if "@" not in email or "." not in email:
        raise Exception("Invalid email format")

    file = open("users.txt", "a")
    file.write("Name: " + name + ", Age: " + str(age) +
                ", Email:" + email + " Track: " + track + "\n")
    file.close()

    file = open("users.txt", "r")
    lines = file.readlines()
    file.close()

    word_count = 0
    for line in lines:
        print(line.upper())
        word_count += len(line.split())

    print("Total words:", word_count)
except Exception as e :
    print("something is wrong:", e )
```

**Task 2:**
Write a Python program that:
1. Asks the user for a **file name**.
2. **Read the file** and display its contents.
3. **Handles exceptions**:
   - If the file does not exist (`FileNotFoundError`), display an error message.
   - Allow the user to **try again** if an error occurs.
4. **If the file is read successfully**, print its contents and exit the program.

### Example Input:
    Enter file name: example.txt
### Example Output:
    Contents of example.txt:
    Hello, this is a sample file.

### Error Example:
    Enter file name: missing.txt
    Error: File 'missing.txt' not found.
    Would you like to try again? (yes/no): yes


```python
while True:
    filename = input("Enter file name: ")

    try:
        with open(filename, "r") as file:
            print(f"Contents of {filename}:")
            print(file.read())
        break

    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
        retry = input("Would you like to try again? (yes/no): ").lower()
        if retry != "yes":
            break
```
