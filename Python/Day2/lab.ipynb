# Strings and Lists in python

1-	Fill an array of 5 elements from the user, Sort it in descending
and ascending orders then display the output

---

```python
arr = []
for i in range(5):
    num = int(input(f"Enter number {i+1}: "))
    arr.append(num)

asc = sorted(arr)
print("Ascending order:", asc)

desc = sorted(arr, reverse=True)
print("Descending order:", desc)
```

3- Write a program that remove all vowels from the input word and
generate a brief version of it

---

```python
word = input("Enter a word: ")

vowels = "aeiouAEIOU"

result = ""
for charchter in word:
    if charchter not in vowels:  
        result += charchter

print("Word without vowels:", result)
```

7-  Write a Python script to:
- Take a string input from the user.
- Convert the string to uppercase.
- Count the number of vowels in the string.
- Reverse the string.
- Check if the string is a **palindrome**
--  example: madam

```python
s = input("Enter a string: ")

s_upper = s.upper()
print("Uppercase:", s_upper)

vowels = "AEIOU"
count = 0
for ch in s_upper:
    if ch in vowels:
        count += 1
print("Vowels count:", count)

for ch in s:
    s_rev = ch + s_rev
print("Reversed string:", s_rev)

if s.lower() == s.lower()[::-1]:
    print("It is a palindrome")
else:
    print("Not a palindrome")
```

# Sets in python

### 1- Create a set of the first five even numbers. Write a script to:

- Add a new number to the set.
- Remove a number from the set.
- Check if a specific number exists in the set.
- Perform a union and intersection with another set `{2, 4, 6, 8, 10}`.


```python
s = {2,4,6,8,10}

s.add(12)
print("After adding 12:", s)

s.remove(4)
print("After removing 4:", s)

if 6 in s:
    print("6 exists in set")

other = {2,4,6,8,10}
union = s.union(other)
intersection = s.intersection(other)
print("Union:", union)
print("Intersection:", intersection)
```

### 2- Unique Visitors Tracker
Create a set to track unique website visitors. Write a script to:

- Add new visitors to the set.
- Check if a visitor has already visited.
- Remove a visitor from the set.


```python
visitors = set()

visitors.add("Alice")
visitors.add("Bob")

if "Alice" in visitors:
    print("Alice already visited")

visitors.remove("Bob")
print("Visitors:", visitors)
```

### 3- Set-Based Password Security Checker

## Task:
Implement a function that checks if a password is weak by comparing it to a predefined set of common passwords.

**Common passwords:**
- `"123456"`, `"password"`, `"123456789"`, `"12345"`, `"12345678"`, `"qwerty"`
- `"1234567"`, `"111111"`, `"123123"`, `"abc123"`, `"password1"`, `"1234"`


```python
common_passwords = {"123456","password","123456789","12345","12345678","qwerty","1234567","111111","123123","abc123","password1","1234"}

def check_password(pwd):
    if pwd in common_passwords:
        print("Weak password")
    else:
        print("Strong password")

password = input("Enter your password: ")
check_password(password)
```

## **Task 2: Tuple-Based Stock Price Analysis**
Given a tuple of stock prices for a week, write a script to:

- Find the highest price.
- Find the lowest price.
- Calculate the average price.
    -  `Output`:
    -	Highest Price: 162
    -	Lowest Price: 145
    -	Average Price: 154.00


```python
fruits = ("apple","banana","orange","grape","mango")
print("First fruit:", fruits[0])
print("Last fruit:", fruits[-1])
print("First three fruits:", fruits[:3])
print("Count of apple:", fruits.count("apple"))
try:
    fruits[0] = "blabla"
except TypeError:
    print("Cannot modify a tuple!")
```

```python
prices = (145,150,155,160,162,148,150)
print("Highest Price:", max(prices))
print("Lowest Price:", min(prices))
average = sum(prices)/len(prices)
print("Average Price:", round(average,2))
```
